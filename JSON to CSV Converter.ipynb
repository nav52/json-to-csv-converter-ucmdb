# JSON to CSV Converter

### Converts UCMDB REST API's JSON output into required CSV.

#### Import required modules

```python
import json
import pandas as pd
from progress.bar import IncrementalBar
```

#### Get the json file and load it

```python
file_name = input("Enter the filename without extension: ")
with open(file_name+'.json', encoding='ansi') as json_data:
    output_data = json.load(json_data)
```

```python
ci_data = output_data['cis']
```

```python
ci_data
```

```python
bar = IncrementalBar('Progress', max = len(ci_data))
```

#### Set up the empty data frame with column headers

```python
column_names = ['Name', 'Vendor', 'Version', 'Type']

ci_list = pd.DataFrame(columns=column_names)
ci_list
```

#### Fill the DataFrame with the appropriate values

```python
for data in ci_data:
    ci_name = data['properties']['name']
    ci_vendor= data['properties']['discovered_vendor'] or "None"
    ci_version= data['properties']['version'] or "None"
    ci_software_type= data['properties']['software_type'] or "None"
    
    ci_list = ci_list.append({'Name':ci_name,
                             'Vendor':ci_vendor,
                             'Version': ci_version,
                             'Type':ci_software_type}, ignore_index=True)
    bar.next()
    
bar.finish()
```

#### Write to a csv file

```python
ci_list.to_csv(file_name+'.csv', index=False, encoding='utf-8')
```
